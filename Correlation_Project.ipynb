## 5. Testing Correlations

```python
movies.corr()
#there are different methods we can use by adding method ='x' where x is pearson(default), kendall, spearman
```

```python
#let's make a visual view of the correlations

correlation_view = movies.corr(method='pearson')
sns.heatmap(correlation_view, annot=True)

plt.title('Correlation View of Numeric Columns from Movies DataFrame')

plt.show()
```

### A) Budget vs Gross

### B) Company vs Gross 

```python
movies_numeric1 = movies
```

```python
#unfortuneately company is not a numeric field so we will have to fix this first
for i in movies_numeric1.columns:
    if(movies_numeric1[i].dtype == 'object'):
        movies_numeric1[i] = movies_numeric1[i].astype('category')
        movies_numeric1[i]= movies_numeric1[i].cat.codes
        
        
movies_numeric1
```

```python
#now we can look at the correlation for every single field 

correlation_view = movies_numeric1

fig, ax = plt.subplots(figsize=(15, 15))
correlation_view = movies.corr(method='pearson')
sns.heatmap(correlation_view, annot=True)


plt.title('Correlation View')

plt.show()
```

```python
#different way of doing this...

correlation_2 = movies_numeric1.corr()
corr_pairs = correlation_2.unstack()
pd.options.display.max_rows = 267 #need this in order to view all of the results 
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) >.5]
high_corr

#Votes and Budget has the highest correlation to gross earnings 
```
